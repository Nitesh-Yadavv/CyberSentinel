# KDD Cup 1999 Dataset - Column Descriptions

---

## Dataset Structure

The dataset comprises various features describing TCP/IP connections, categorized into five groups to capture different aspects of network traffic.

---

### 1. Basic Features of Individual TCP Connections

These features describe fundamental properties of each network connection.

| Column          | Description                                                  |
| :-------------- | :----------------------------------------------------------- |
| `duration`      | Length of the connection (in seconds)                        |
| `protocol_type` | Type of protocol used (`tcp`, `udp`, `icmp`)                 |
| `service`       | Network service on the destination (`http`, `ftp`, etc.)     |
| `flag`          | Normal or error status of the connection (`SF`, `REJ`, etc.) |
| `src_bytes`     | Number of data bytes from source to destination              |
| `dst_bytes`     | Number of data bytes from destination to source              |
| `land`          | 1 if connection is from/to the same host/port; else 0        |

---

###  2. Content Features Within a Connection

These features are derived from the content inspection of the data exchanged during a connection.

| Column               | Description                                                     |
| :------------------- | :-------------------------------------------------------------- |
| `wrong_fragment`     | Number of wrong fragments                                       |
| `urgent`             | Number of urgent packets                                        |
| `hot`                | Number of "hot" indicators (e.g., access to system directories) |
| `num_failed_logins`  | Number of failed login attempts                                 |
| `logged_in`          | 1 if successfully logged in; else 0                             |
| `num_compromised`    | Number of compromised conditions                                |
| `root_shell`         | 1 if root shell is obtained; else 0                             |
| `su_attempted`       | 1 if "su root" command attempted; else 0                        |
| `num_root`           | Number of root accesses                                         |
| `num_file_creations` | Number of file creation operations                              |
| `num_shells`         | Number of shell prompts invoked                                 |
| `num_access_files`   | Number of access control files accessed                         |
| `num_outbound_cmds`  | Number of outbound commands (always 0 in this dataset)          |
| `is_host_login`      | 1 if login belongs to the host list; else 0                     |
| `is_guest_login`     | 1 if guest login; else 0                                        |

---

###  3. Traffic Features Based on Connections in the Last 2 Seconds

These features are calculated over a time window (specifically, the last 2 seconds) to capture short-term behavioral patterns related to multiple connections.

| Column               | Description                                                     |
| :------------------- | :-------------------------------------------------------------- |
| `count`              | Number of connections to the same host in the past 2 seconds    |
| `srv_count`          | Number of connections to the same service in the past 2 seconds |
| `serror_rate`        | % of connections that have "SYN" errors                         |
| `srv_serror_rate`    | % of connections with "SYN" errors to the same service          |
| `rerror_rate`        | % of connections with "REJ" errors                              |
| `srv_rerror_rate`    | % of connections with "REJ" errors to the same service          |
| `same_srv_rate`      | % of connections to the same service                            |
| `diff_srv_rate`      | % of connections to different services                          |
| `srv_diff_host_rate` | % of connections to different hosts                             |

---

### 4. Traffic Features Based on Destination Host

These features capture long-term behavioral patterns related to connections directed towards the same destination host.

| Column                        | Description                                              |
| :---------------------------- | :------------------------------------------------------- |
| `dst_host_count`              | Number of connections to the same destination host                   |
| `dst_host_srv_count`          | Number of connections to the same service on the destination host |
| `dst_host_same_srv_rate`      | % of connections to the same service on the destination host                     |
| `dst_host_diff_srv_rate`      | % of connections to different services on the destination host                   |
| `dst_host_same_src_port_rate` | % of connections from the same source port to the destination host               |
| `dst_host_srv_diff_host_rate` | % of connections to different hosts for the same service on the destination host |
| `dst_host_serror_rate`        | % of connections to the destination host that have SYN errors                    |
| `dst_host_srv_serror_rate`    | % of connections to the same service on the destination host with SYN errors     |
| `dst_host_rerror_rate`        | % of connections to the destination host with REJ errors                         |
| `dst_host_srv_rerror_rate`    | % of connections to the same service on the destination host with REJ errors     |

---

### 5. Label

This is the target variable indicating the nature of the connection.

| Column  | Description                                                                  |
| :------ | :--------------------------------------------------------------------------- |
| `label` | Type of connection: `normal.` or an attack name (e.g., `smurf.`, `neptune.`) |

---

This README aims to provide a clear and concise reference for anyone working with the KDD Cup 1999 dataset.

```python
import pandas as pd

# Full list of column names based on kddcup.names file
column_names = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes',
    'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in',
    'num_compromised', 'root_shell', 'su_attempted', 'num_root',
    'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds',
    'is_host_login', 'is_guest_login', 'count', 'srv_count',
    'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate',
    'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count',
    'dst_host_srv_count', 'dst_host_same_srv_rate', 'dst_host_diff_srv_rate',
    'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate',
    'dst_host_serror_rate', 'dst_host_srv_serror_rate',
    'dst_host_rerror_rate', 'dst_host_srv_rerror_rate', 'label'
]

# Path to raw data
data_path = '../data/raw/kddcup.data_10_percent_corrected'


# Assign column names
df = pd.read_csv(data_path, names=column_names)

df
```

# Drop the `label` column for unsupervised learning 

```python
# Store labels for evaluation
labels = df['label'].copy()

# Drop the label column for training
df.drop(columns=['label'], inplace=True)

df.columns.tolist()
```

```python
df.shape
```

```python
# Class Distribution
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize=(10, 6))
sns.countplot(y=labels, order=labels.value_counts().index, palette="viridis")
plt.title("Class Distribution (Attack Types)")
plt.xlabel("Count")
plt.ylabel("Attack Type")
plt.tight_layout()
plt.show()
```

```python
# Protocol Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='protocol_type', palette='Set2')
plt.title("Protocol Type Distribution")
plt.xlabel("Protocol")
plt.ylabel("Count")
plt.show()
```

```python
# Flag Distribution
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='flag', order=df['flag'].value_counts().index, palette='coolwarm')
plt.title("Flag Field Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Numerical Feature Boxplots
numerical_features = ['src_bytes', 'dst_bytes', 'count', 'srv_count']
df_temp = df.copy()
df_temp['label'] = labels

for col in numerical_features:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x='label', y=col, data=df_temp[df_temp['label'].isin(['normal.', 'smurf.', 'neptune.'])], palette='mako')
    plt.title(f"Distribution of {col} across Classes")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Correlation Heatmap
import numpy as np
plt.figure(figsize=(20,18))
corr = df.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0, annot=True, fmt=".2f",
            linewidths=0.5, cbar_kws={"shrink": .5})
plt.title("Feature Correlation Heatmap with All Columns")
plt.tight_layout()
plt.show()
```

```python
# Encode label: normal = 0, attack = 1
df_temp['binary_label'] = df_temp['label'].apply(lambda x: 0 if x == 'normal.' else 1)

# Correlation of numeric features with binary label
corr_with_label = df_temp.select_dtypes(include='number').corr()['binary_label'].drop('binary_label').sort_values(ascending=False)

# Plot correlation with label
plt.figure(figsize=(10, 8))
sns.barplot(x=corr_with_label.values, y=corr_with_label.index, palette='rocket')
plt.title("Correlation of Features with Label (0=normal, 1=attack)")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
# Select top N most correlated features with binary label
top_n = 10  # you can change this to any number
top_features = corr_with_label.abs().sort_values(ascending=False).head(top_n)

print(f"Top {top_n} features most correlated with the label:")
print(top_features)
```
